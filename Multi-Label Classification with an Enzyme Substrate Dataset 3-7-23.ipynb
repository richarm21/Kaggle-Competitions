## Importing target variable

## Model Building using Logistic Regression

* ### Import model

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
```

* ### Predicting variable EC1

```python
# Defining variables
xtrain = df_train.iloc[:,:-6]
ytrain_EC1 = df_train.iloc[:,-6]
xtest = df_test
ytest_EC1 = target.iloc[:,1]
```

```python
# Fitting data
log_model.fit(xtrain, ytrain_EC1)
```

```python
# Prediction
ypredtest_EC1 = log_model.predict(xtest)
ypredtest_EC1
```

```python
# Finding probability
ypredprob_EC1 = log_model.predict_proba(xtest)
ypredprob_EC1
```

* ### Predicting variable EC2

```python
# Defining variables
xtrain = df_train.iloc[:,:-6]
ytrain_EC2 = df_train.iloc[:,-5]
xtest = df_test
ytest_EC2 = target.iloc[:,2]
```

```python
# Fitting data
log_model.fit(xtrain, ytrain_EC2)
```

```python
# Prediction
ypredtest_EC2 = log_model.predict(xtest)
ypredtest_EC2
```

```python
# Finding probability
ypredprob_EC2 = log_model.predict_proba(xtest)
ypredprob_EC2
```

```python
#Saving predictions in sample submission file
probabilities = pd.DataFrame(ypredprob_EC1[:,1])
file11 = probabilities.to_csv("Final Submission using Logistic Regression 3-7-23.csv", index = False)

file11 = pd.read_csv("Final Submission using Logistic Regression 3-7-23.csv")
file11 = pd.concat([target['id'],probabilities,pd.DataFrame(ypredprob_EC2[:,1])], axis = 1)
file11.columns = ["id", "EC1","EC2"]
file11.to_csv("Final Submission using Logistic Regression 3-7-23.csv", index = False)
```
